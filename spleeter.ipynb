# create the training data

```python
import os
import pandas as pd
import ffmpeg
```

```python
import wave
import contextlib
# fname = r'D:\voice\dnr_fix\tr\89839\mix.wav'
# get_size(fname)
```

```python
def get_size(fname):
    with contextlib.closing(wave.open(fname,'r')) as f:
        frames = f.getnframes()
        rate = f.getframerate()
        duration = frames / float(rate)
        return duration
```

```python
def convert_to_wav(file_name):
    filename =  f"D:\\voice\\dnr\\{file_name}"
    actual_filename = filename.replace("dnr", "dnr_fix")
    os.makedirs(os.path.dirname(actual_filename), exist_ok=True)
    command = 'ffmpeg -i {} -f wav -bitexact -acodec pcm_s16le -ar 44100 -ac 2 -y {}'.format(filename, actual_filename)
    os.system(command)
    print(actual_filename)
```

# convert file

```python
training_folder = r"D:\voice\dnr\cv"
for folder in os.listdir(training_folder):
    sub_folder = f"{training_folder}\\{folder}"
    if os.path.isdir(sub_folder):
        for file in os.listdir(sub_folder):
            if file.endswith("wav"):
                convert_to_wav(f"cv\\{folder}\\{file}")
```

```python
training_folder = r"D:\voice\dnr_fix\tr"
for folder in os.listdir(training_folder):
    sub_folder = f"{training_folder}\\{folder}"
    if os.path.isdir(sub_folder):
        for file in os.listdir(sub_folder):
            if file == "mix.wav":
                # f"{sub_folder}\\mix.wav"
                os.rename(f"{sub_folder}\\mix.wav", f'{sub_folder}\\mixture.wav')
```
